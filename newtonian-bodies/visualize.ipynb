```python
import itertools
import pathlib

import polars as pl
import matplotlib.pyplot as plt
```

```python
def plot_trajectories(df: pl.DataFrame, names: tuple[str, ]):
    """
    Plots the 3D trajectories for all celestial bodies in the DataFrame.
    """
    
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Create a color cycle for the plots
    colors = itertools.cycle(['blue', 'red', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan'])

    # Add the Sun at the center for context
    ax.scatter([0], [0], [0], color='yellow', s=250, label='Sun (Origin)')

    for name, color in zip(names, colors):
        # 1. Correctly filter the *original* DataFrame for each body
        body_df = df.filter(pl.col("name") == name)

        # Extract position data
        pos_x = body_df["pos_x"].to_numpy()
        pos_y = body_df["pos_y"].to_numpy()
        pos_z = body_df["pos_z"].to_numpy()

        # 2. Plot the trajectory with a unique color and label
        ax.plot(pos_x, pos_y, pos_z, label=f"{name}'s Trajectory", color=color)

    # Set labels, title, and other plot properties outside the loop
    ax.set_xlabel("X Position (meters)")
    ax.set_ylabel("Y Position (meters)")
    ax.set_zlabel("Z Position (meters)")
    ax.set_title("Trajectories of Newtonian Bodies")
    ax.set_aspect('equal', adjustable='box')
    ax.legend()
    
    plt.show()
```

```python
df = pl.read_parquet("newtonian.parquet")
```

```python
plot_trajectories(df, names=("Sun", "Earth"))
```
